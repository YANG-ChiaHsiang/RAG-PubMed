```python
import os
import pandas as pd

def merge_csv_files(csv_folder, output_file):
    first_file = True
    file_count = 0
    
    with open(output_file, 'w', newline='', encoding='utf-8') as out_file:
        for file in os.listdir(csv_folder):
            if file.endswith(".csv"):
                file_path = os.path.join(csv_folder, file)
                try:
                    df = pd.read_csv(file_path)
                    df = df.dropna(subset=["Title", "Abstract"])  # 過濾掉 Title 或 Abstract 為 NaN 的行
                    file_count += 1
                    
                    df.to_csv(out_file, mode='a', index=False, header=first_file)
                    first_file = False  # 之後的文件不寫入標頭
                    print(f"已處理第 {file_count} 個檔案: {file}")
                except Exception as e:
                    print(f"讀取 {file} 時發生錯誤: {e}")
    
    print(f"合併後的 CSV 已儲存至 {output_file}")

if __name__ == "__main__":
    folder_path = "../data/pubmed_baseline/csv"  # 指定 csv 資料夾
    output_csv = "../data/pubmed_baseline/merged_output.csv"  # 輸出檔案名稱
    merge_csv_files(folder_path, output_csv)
```
