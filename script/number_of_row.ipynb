```python
import os
import pandas as pd

def count_total_rows(csv_folder):
    total_rows = 0
    
    # 遍歷 csv/ 目錄下的所有 .csv 檔案
    for file in os.listdir(csv_folder):
        print(f"total_rows: {total_rows}")
        if file.endswith(".csv"):
            file_path = os.path.join(csv_folder, file)
            try:
                df = pd.read_csv(file_path)
                total_rows += len(df)
            except Exception as e:
                print(f"讀取 {file} 時發生錯誤: {e}")
    
    return total_rows
```

```python
folder_path = "../data/pubmed_baseline/csv"  # 指定 csv 資料夾
total_rows = count_total_rows(folder_path)
print(f"csv/ 資料夾內所有 .csv 檔案的總行數（不含標頭）: {total_rows}")
```
